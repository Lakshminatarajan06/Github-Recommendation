```python
import requests
from requests.utils import quote

def search_users(query, page=1, per_page=100):
    headers = {
        'Authorization': token,  
        'Accept': 'application/vnd.github.v3+json'
    }
    
    encoded_query = quote(query)
    url = f'https://api.github.com/search/users?q={encoded_query}&page={page}&per_page={per_page}'
    
    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Raise an HTTPError for bad responses (4xx and 5xx)
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"Failed to search users: {e}")
        return None

# Function to get usernames based on the search query
def get_usernames(query, max_usernames=100):
    usernames = []
    page = 1
    
    while len(usernames) < max_usernames:
        result = search_users(query, page)
        if result and 'items' in result:
            usernames.extend([item['login'] for item in result['items']])
            if len(result['items']) == 0:
                break  # No more results
        else:
            break  # Stop if the request fails or no more results
        page += 1
    
    return usernames[:max_usernames]

# Example usage
query = 'location:India'  # Example query, adjust as needed
usernames = get_usernames(query)
print(usernames)
```

```python
from datetime import datetime
from collections import defaultdict

# Your GitHub Personal Access Token
TOKEN = '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'

# GitHub API base URL
BASE_URL = 'https://api.github.com'

# Headers for authentication
HEADERS = {
    'Authorization': f'token {TOKEN}',
    'Accept': 'application/vnd.github.v3+json'
}

def get_user_data(username):
    url = f'{BASE_URL}/users/{username}'
    response = requests.get(url, headers=HEADERS)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error fetching user data: {response.json()}")
        return {}

def get_following(username):
    url = f'{BASE_URL}/users/{username}/following'
    response = requests.get(url, headers=HEADERS)
    if response.status_code == 200:
        return [user['login'] for user in response.json()]
    else:
        print(f"Error fetching following list: {response.json()}")
        return []

def get_followers(username):
    url = f'{BASE_URL}/users/{username}/followers'
    response = requests.get(url, headers=HEADERS)
    if response.status_code == 200:
        return [user['login'] for user in response.json()]
    else:
        print(f"Error fetching followers list: {response.json()}")
        return []

def get_repos(username):
    url = f'{BASE_URL}/users/{username}/repos'
    response = requests.get(url, headers=HEADERS)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error fetching repositories: {response.json()}")
        return []

def get_languages(repo_url):
    url = f'{repo_url}/languages'
    response = requests.get(url, headers=HEADERS)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Error fetching languages: {response.json()}")
        return {}

def get_stars(repo):
    return repo.get('stargazers_count', 0)

def get_commits(repo_full_name):
    commits_url = f'{BASE_URL}/repos/{repo_full_name}/commits'
    commits = []
    page = 1
    while True:
        response = requests.get(f'{commits_url}?page={page}', headers=HEADERS)
        if response.status_code == 200:
            data = response.json()
            if not data:
                break
            commits.extend(data)
            page += 1
        else:
            print(f"Error fetching commits: {response.json()}")
            break
    return commits

def compile_commits_per_quarter(username):
    repos = get_repos(username)
    commits_per_quarter = defaultdict(int)

    for repo in repos:
        repo_full_name = repo['full_name']
        commits = get_commits(repo_full_name)
        
        for commit in commits:
            commit_date = commit['commit']['committer']['date']
            commit_datetime = datetime.strptime(commit_date, '%Y-%m-%dT%H:%M:%SZ')
            quarter = (commit_datetime.year, (commit_datetime.month - 1) // 3 + 1)
            commits_per_quarter[quarter] += 1

    return commits_per_quarter

def compile_user_data(username):
    user_data = get_user_data(username)
    
    if not user_data:
        return {}

    following_list = get_following(username)
    followers_list = get_followers(username)
    repos = get_repos(username)
    
    languages = {}
    repo_count_per_language = {}
    stars_per_language = {}
    commits_per_language = {}
    stars_per_repo = {}
    commits_per_repo = {}

    for repo in repos:
        repo_name = repo['name']
        stars = get_stars(repo)
        commits = get_commits(repo['full_name'])
        repo_languages = get_languages(repo['url'])
        
        stars_per_repo[repo_name] = stars
        commits_per_repo[repo_name] = len(commits)
        
        for lang, count in repo_languages.items():
            if lang not in languages:
                languages[lang] = 0
                repo_count_per_language[lang] = 0
                stars_per_language[lang] = 0
                commits_per_language[lang] = 0
            
            languages[lang] += count
            repo_count_per_language[lang] += 1
            stars_per_language[lang] += stars
            commits_per_language[lang] += len(commits)
    
    starred_repos = sum(stars_per_language.values())
    total_commits = sum(commits_per_language.values())

    commits_per_quarter = compile_commits_per_quarter(username)
    
    return {
        'login': user_data.get('login'),
        'username': user_data.get('login'),
        'created_on': user_data.get('created_at'),
        'updated_on': user_data.get('updated_at'),
        'name': user_data.get('name'),
        'bio': user_data.get('bio'),
        'following_count': user_data.get('following'),
        'following_list': following_list,
        'follower_count': user_data.get('followers'),
        'follower_list': followers_list,
        'avatar': user_data.get('avatar_url'),
        'url': user_data.get('html_url'),
        'languages': languages,
        'public_repositories': user_data.get('public_repos'),
        'repository_per_language': repo_count_per_language,
        'stars_per_repository': stars_per_repo,
        'commits_per_repository': commits_per_repo,
        'total_commits': total_commits,
        'commits_per_language': commits_per_language,
        'stars_per_language': stars_per_language,
        'starred_repositories': starred_repos,
        'commits_per_quarter': commits_per_quarter
    }

# Example usage
username = 'nethajinirmal13'
user_data= compile_user_data(username)
pprint(user_data)
```
